```python
# Programa II: 

# Baixar os dados da API iTunes sobre as músicas de um determinado grupo ou artista e gerar uma planilha Excel com esses dados:

import pandas as pd
import requests
import json

base_site = "https://itunes.apple.com/search"
r = requests.get(base_site, params = {"term": "The Beatles", "country": "us"})
info = r.json()
songs_df = pd.DataFrame(info["results"])
songs_df.to_excel("musicas_beatles_info.xlsx")
songs_df
```
