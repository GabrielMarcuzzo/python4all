```python
# Programa III:

# Gráfico de linha utilizando a biblioteca Matplotlib.

import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,3,11,14]
plt.title("Gráfico de linhas criado com Python")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.plot(x, y)
plt.show()
```
