```python
# Análise de Cluster - algoritmo k-means não supervisionado

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans

data = pd.read_excel("D:/IP_unicornios_1.xlsx")

plt.scatter(data['World'],data['AL'])
plt.xlabel('Trademark Registrations - US and Europe')
plt.ylabel('Latin America')

x = data.copy()

kmeans = KMeans(3)
kmeans.fit(x)

clusters = x.copy()
clusters['cluster_pred']=kmeans.fit_predict(x)

plt.scatter(clusters['World'],clusters['AL'],c=clusters['cluster_pred'],cmap='rainbow')
plt.xlabel('Trademark Registrations - US and Europe')
plt.ylabel('Trademark Registrations - Latin America')

```
