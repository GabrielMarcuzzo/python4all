```python
# Programa I:

# Acessar a API do Yahoo Finance para Plotar Grfico de Volatilidade de Aes.

import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt

tickers = ["NFLX"]
stock_prices_br = pd.DataFrame()
for t in tickers:
     stock_prices_br[t] = wb.DataReader(t, data_source = "yahoo", start ="2020-1-1")["Adj Close"]
stock_prices_br.plot(figsize = (15, 6))
plt.title("Preo das Aes da Netflix: 2020-Presente")
plt.show
```
