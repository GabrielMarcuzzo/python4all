```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt

tickers = ["TSLA", "AMZN", "GOOGL", "FB", "NFLX", "MELI", "STNE"]
stock_prices_br = pd.DataFrame()
for t in tickers:
	stock_prices_br[t] = wb.DataReader(t, data_source = "yahoo", start ="2021-1-1")["Adj Close"]

(stock_prices_br/ stock_prices_br.iloc[0] * 100).plot(figsize = (15, 6));
plt.show
```
