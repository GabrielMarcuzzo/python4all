```python
import matplotlib.pyplot as plt
import yfinance as yf

petr = yf.Ticker("PETR4.SA")

data = petr.history(period="max") # onde 'period' pode ser: 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 10y, ytd, ou max para todo o historico

data = petr.history(start = "2000-01-01", end = "2021-09-17")

titulo = "Dividendos Pagos pela Petrobras (PETR4) nos Últimos 20 anos"

plt.title(titulo)

data.Dividends.plot();
```
