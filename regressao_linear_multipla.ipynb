```python
# Passo 3: Regressão linear múltipla

import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt

data = pd.read_excel("D:/planilha_venda_casas.xlsx")

X = data[["Terreno", "Quartos", "Construcao"]]
Y = data["Preco"]

X1 = sm.add_constant(X)
reg = sm.OLS(Y, X1).fit()

reg.summary()
```
